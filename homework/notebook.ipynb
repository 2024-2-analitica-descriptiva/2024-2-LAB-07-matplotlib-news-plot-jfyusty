```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargamos el archivo CSV
file_path = '../files/input/news.csv'
df = pd.read_csv(file_path, index_col=0)
df.head()
```

```python
plt.figure()

for col in df.columns:
    plt.plot(df[col], label=col)

plt.title("How people get their news", fontsize=16)
plt.legend(loc="upper right")
plt.show
```

```python
plt.figure(figsize=(10, 8))

colors = {
    'Television': 'dimgray',
    'Newspaper': 'grey',
    'Internet': 'tab:blue',
    'Radio': 'lightgrey',
}

zorder = {
    'Television': 1,
    'Newspaper': 1,
    'Internet': 2,
    'Radio': 1,
}

linewidths = {
    'Television': 2,
    'Newspaper': 2,
    'Internet': 4,
    'Radio': 2,
}

for col in df.columns:
    plt.plot(df[col],
             color=colors[col],
             label=col,
             zorder=zorder[col],
             linewidth=linewidths[col],
             )

plt.title("How people get their news", fontsize=16)
plt.text(x=0.5, y=1, s="Una proporcin cada vez mayor cita Internet como su principal fuente de noticias",
             ha='center', va='center', transform=plt.gca().transAxes, fontsize=8)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)

for col in df.columns:
    first_year = df.index[0]
    plt.scatter(
        x=first_year,
        y=df[col][first_year],
        color=colors[col],
        zorder=zorder[col],
    )

    plt.text(
        first_year - 0.2,
        df[col][first_year],
        col +" " +str(df[col][first_year]) + "%",
        ha='right',
        va='center',
        color=colors[col],
    )

    last_year = df.index[-1]
    plt.scatter(
        x=last_year,
        y=df[col][last_year],
        color=colors[col],
    )

    plt.text(
        last_year + 0.2,
        df[col][last_year],
        str(df[col][last_year]) + "%",
        ha='left',
        va='center',
        color=colors[col],
    )

plt.xticks(
    ticks=df.index,
    labels=df.index,
    ha='center',
)

plt.show
```

```python
file_path = '../files/input/news.csv'
df = pd.read_csv(file_path, index_col=0)

plt.figure()  # Ajustar tamao para mejor visualizacin

colors = {
    'Television': 'dimgray',
    'Newspaper': 'grey',
    'Internet': 'tab:blue',
    'Radio': 'lightgrey',
}

zorder = {
    'Television': 3,
    'Newspaper': 2,
    'Internet': 4,
    'Radio': 1,
}

linewidths = {
    'Television': 2,
    'Newspaper': 2,
    'Internet': 3,
    'Radio': 2,
}

markers = {
    'Television': 'o',
    'Newspaper': 'o',
    'Internet': 'o',
    'Radio': 'o',
}

for col in df.columns:
    plt.plot(df.index, df[col], color=colors[col], label=col, zorder=zorder[col],
                linewidth=linewidths[col], marker=markers[col], markersize=8)  # Agregamos marcadores

plt.title("How people get their news", ha='center', va='center', fontsize=16, y=1.05)
plt.text(0.5, 1, "Una proporcin cada vez mayor cita Internet como su principal fuente de noticias",
            ha='center', va='center', transform=plt.gca().transAxes, fontsize=8)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)

plt.xticks(ticks=df.index, labels=df.index, ha='center')

for col in df.columns:
    last_year = df.index[-1]
    plt.text(last_year + 0.2, df[col][last_year], f'{df[col][last_year]}%',
                ha='left', va='center', color=colors[col], fontsize=12)
```
